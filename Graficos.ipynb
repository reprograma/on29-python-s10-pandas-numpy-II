Utilizar a tabela de dados do clima de seu estado crie 3 gráficos com os dados fornecidos pela tabela, os gráficos podem ser de qualquer dado, qualquer tipo e de qualquer amostragem



```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_sample_CE.csv',  sep=';' , encoding='latin-1', on_bad_lines='skip')
```

```python
df
```

```python
print(df.columns)
```

```python
import numpy as np
```

```python
df = df.dropna(subset=['TEMPERATURA DO AR - BULBO SECO. HORARIA (C)', 'VENTO. VELOCIDADE HORARIA (m/s)'])
```

```python
df.to_numpy()
```

```python
temperatura = df['TEMPERATURA DO AR - BULBO SECO. HORARIA (C)'].to_numpy()
```

```python
vento_ms = df['VENTO. VELOCIDADE HORARIA (m/s)'].to_numpy()
```

```python
vento_kh = vento_ms * 3.6
```

```python
vento_kh
```

```python
np.sqrt(vento_kh)
```

```python
sensacao = 33 + (10 * np.sqrt(vento_kh) + 10.45 - vento_kh) * (temperatura - 33) / 22
```

```python
sensacao
```

```python
data = df['Data'].to_numpy()
```

```python
hora = df['Hora UTC'].str.replace('00 UTC', ':00').to_numpy()
```

```python
df_calculado = pd.DataFrame({'Data':data, 'Hora':hora, 'Temperatura':temperatura, 'Vento':vento_ms, 'Sensacao Termica':sensacao})
```

```python
media_hora = df_calculado.groupby('Hora')[['Temperatura', 'Sensacao Termica']].mean()
```

```python
media_hora['Hora'] = media_hora.index
```

```python
media_hora.plot()
```

```python
media_hora['Hora int'] = media_hora.index.str.replace(':00','')
```

```python
media_hora['Hora int'] = pd.to_numeric(media_hora['Hora int'])
```

```python
cond1 = media_hora['Hora int'] >= 8
cond2 = media_hora['Hora int'] <= 18

media_hora[cond1 & cond2][['Temperatura', 'Sensacao Termica']].plot()
```

```python
Grupos = df.groupby('ESTACAO')[["PRECIPITAO TOTAL. HORARIO (mm)","TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)"]].mean()
```

```python
Grupos
```

```python
print(df.columns)
```

```python
fig, axs = plt.subplots(len(Grupos.columns), 1, figsize=(10, 8), sharex=True)
for i, coluna in enumerate(Grupos.columns):
    Grupos[coluna].plot(kind='bar', ax=axs[i])
    axs[i].set_ylabel(coluna)

plt.xlabel('ESTACAO')
plt.tight_layout()
plt.show()
```
