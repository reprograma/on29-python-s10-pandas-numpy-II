```python
import pandas as pd
```

```python
df = pd.read_csv('data_sample_AM.csv', sep=';')
```

```python
df
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

Criação do primeiro gráfico: média da temperatura maior que 30

```python
temperatura = df["TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)"] > 30

filtro_temperatura = df[temperatura]["TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)"].value_counts()

filtro_temperatura.plot.hist(bins=20, edgecolor = "red", color = "pink")

plt.xlabel("Média da Temperatura")
plt.ylabel("Quantidade da Temperatura")
plt.title("Temperatura Máxima na Hora Ant. (AUT) (c)")
```

Criação do segundo gráfico: média de radição por dia

```python
df["Data Hora"] = df["Data"] + " " + df["Hora UTC"].str.replace("00 UTC", " ")
```

```python
media_dia = df[["Data", "RADIACAO GLOBAL (Kj/m2)"]].groupby("Data").mean()
```

```python
media_dia.plot()

plt.xlabel("Indíces de radiação por data")
plt.ylabel("Frequência da radiação")
plt.title("Média da RADIACAO GLOBAL (Kj/m2) ")
```

Criação do terceiro gráfico: comparação de pressão máxima e mínima

```python
df["Data Hora"] = df["Data"] + " " + df["Hora UTC"].str.replace("00 UTC", " ")
```

```python
comparacao_pressao = df[["Data", "PRESSAO ATMOSFERICA MAX.NA HORA ANT. (AUT) (mB)","PRESSAO ATMOSFERICA MIN. NA HORA ANT. (AUT) (mB)" ]].groupby("Data").mean()
```

```python
comparacao_pressao.plot()

plt.xlabel("Indíces de pressão por data")
plt.ylabel("Variação de pressão")
plt.title("Comparação da Pressão Atmosférica Máxima e Mínima por data")
```
