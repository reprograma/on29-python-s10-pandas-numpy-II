Importar bibliotecas: pandas, matplotilib e numpy



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Leitura do arquivo

```python
df = pd.read_csv('data_sample_SP.csv', sep=';', encoding= 'latin-1', on_bad_lines='skip')
```

Print do Dataframe

```python
df
```

```python
df = df.dropna(subset= ['PRECIPITAO TOTAL. HORARIO (mm)'])
df
```

Média de precipitação por estação.

```python
df.groupby('ESTACAO')[['PRECIPITAO TOTAL. HORARIO (mm)']].mean().plot(color= 'purple', figsize=(10, 5))
plt.xlabel('Estação')
plt.ylabel('Precipitação total (mm)')
plt.title('Média da Precipitação por Estação')
```

Selecionar dados de **Piracicaba**, ordenar por data, excluir dados NaN

```python
df_pira = df.loc[df['ESTACAO'] == 'PIRACICABA']
df_pira = df_pira.dropna()
df_pira.sort_values(by= 'Data', ascending = True)
```

Gráfico da média de variação de temperatura ao longo dia da estação de Piracicaba.



```python
df_pira.groupby('Hora UTC')[['TEMPERATURA DO AR - BULBO SECO. HORARIA (C)']].mean().plot(color= 'purple', figsize=(10, 5))
plt.xlabel('Hora UTC')
plt.ylabel('Temperatura do ar (°C)')
plt.title('Média da Temperatura do Ar por hora na Estação Piracicaba')
```

Gráfico das máximas de pressão atmosférica agrupados por dia da estação de Piracicaba.

```python
df_pira.groupby('Data')[['PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO. HORARIA (mB)']].max().plot(figsize=(10, 5))
plt.xlabel('Data')
plt.ylabel('Pressão Atmosférica Máxima (mB)')
plt.title('Máxima Pressão Atmosférica por dia na Estação Piracicaba')
```

```python
df_sanca = df.loc[df['ESTACAO'] == 'SAO CARLOS']
df_sanca
```

Gráfico da pressão atmosférica máxima por hora na Estação São Carlos.

```python
df_sanca.groupby('Hora UTC')[['PRESSAO ATMOSFERICA MAX.NA HORA ANT. (AUT) (mB)']].mean().plot(figsize=(10, 5))
plt.xlabel('Pressão Atmosférica Máxima (mB)')
plt.ylabel('Hora UTC')
plt.title('Média da Pressão Atmosférica Máxima por hora na Estação São Carlos')
```
