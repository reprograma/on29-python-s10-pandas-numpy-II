Importar bibliotecas: pandas, matplotilib e numpy



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Leitura do arquivo

```python
df = pd.read_csv('data_sample_SP.csv', sep=';', encoding= 'latin-1', on_bad_lines='skip')
```

Print do Dataframe

```python
df
```

```python
df = df.dropna(subset= ['PRECIPITAO TOTAL. HORARIO (mm)'])
df
```

MÃ©dia de precipitaÃ§Ã£o por estaÃ§Ã£o.

```python
df.groupby('ESTACAO')[['PRECIPITAO TOTAL. HORARIO (mm)']].mean().plot(color= 'purple', figsize=(10, 5))
plt.xlabel('EstaÃ§Ã£o')
plt.ylabel('PrecipitaÃ§Ã£o total (mm)')
plt.title('MÃ©dia da PrecipitaÃ§Ã£o por EstaÃ§Ã£o')
```

Selecionar dados de **Piracicaba**, ordenar por data, excluir dados NaN

```python
df_pira = df.loc[df['ESTACAO'] == 'PIRACICABA']
df_pira = df_pira.dropna()
df_pira.sort_values(by= 'Data', ascending = True)
```

GrÃ¡fico da mÃ©dia de variaÃ§Ã£o de temperatura ao longo dia da estaÃ§Ã£o de Piracicaba.



```python
df_pira.groupby('Hora UTC')[['TEMPERATURA DO AR - BULBO SECO. HORARIA (C)']].mean().plot(color= 'purple', figsize=(10, 5))
plt.xlabel('Hora UTC')
plt.ylabel('Temperatura do ar (Â°C)')
plt.title('MÃ©dia da Temperatura do Ar por hora na EstaÃ§Ã£o Piracicaba')
```

GrÃ¡fico das mÃ¡ximas de pressÃ£o atmosfÃ©rica agrupados por dia da estaÃ§Ã£o de Piracicaba.

```python
df_pira.groupby('Data')[['PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO. HORARIA (mB)']].max().plot(figsize=(10, 5))
plt.xlabel('Data')
plt.ylabel('PressÃ£o AtmosfÃ©rica MÃ¡xima (mB)')
plt.title('MÃ¡xima PressÃ£o AtmosfÃ©rica por dia na EstaÃ§Ã£o Piracicaba')
```
