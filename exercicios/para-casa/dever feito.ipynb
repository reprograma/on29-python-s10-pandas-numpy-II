```python
import pandas as pd
import numpy as np

# Lendo o arquivo CSV
df = pd.read_csv('data_sample_GO.csv', sep=';')

# Convertendo a coluna 'Data' para o formato datetime
df['Data'] = pd.to_datetime(df['Data'], format='%Y/%m/%d')

# Selecionando colunas  e removendo nulos
df = df[['Data', 'TEMPERATURA DO AR - BULBO SECO. HORARIA (C)']].dropna()

# Verificando as primeiras linhas do DataFrame
df.head()
```

```python
# Agrupando por mês e calcular a média
df['Month'] = df['Data'].dt.to_period('M')
grouped = df.groupby('Month').mean()

# Verificando os dados agrupados
grouped.head()
```

```python
import matplotlib.pyplot as plt

# Configurações gerais do gráficos 
plt.style.use('seaborn-darkgrid')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10

# Convertendo o índice para datetime
months = grouped.index.to_timestamp()

# Gráfico Temperatura do Ar ao longo do tempo média mensal
plt.figure()
plt.plot(months, grouped['TEMPERATURA DO AR - BULBO SECO. HORARIA (C)'], label='Temperatura do Ar', color='tab:red', linewidth=2)
plt.title('Temperatura do Ar ao longo do tempo (média mensal)')
plt.xlabel('Data')
plt.ylabel('Temperatura (C)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Lendo o arquivo CSV
df = pd.read_csv('data_sample_GO.csv', sep=';')

# Convertendo a coluna 'Data' para o formato datetime
df['Data'] = pd.to_datetime(df['Data'], format='%Y/%m/%d')

# Selecionando colunas  e removendo linhas nulas
df = df[['Data', 'TEMPERATURA DO AR - BULBO SECO. HORARIA (C)', 'PRECIPITAO TOTAL. HORARIO (mm)']].dropna()

# Verificando as primeiras linhas do DataFrame
df.head()
```

```python
# Agrupando por mês e calculando a média
df['Month'] = df['Data'].dt.to_period('M')
grouped = df.groupby('Month').mean()

# Verificando 
grouped.head()
```

```python
import matplotlib.pyplot as plt

# Configurações do gráficos 
plt.style.use('seaborn-darkgrid')
plt.rcParams['figure.figsize'] = (12, 12)  # Ajustar o tamanho da figura
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10

# Convertendo o índice para datetime
months = grouped.index.to_timestamp()

# Criar subplots
fig, ax = plt.subplots(2, 1, sharex=True)

# Gráfico 1: Temperatura do Ar ao longo do tempo (média mensal)
ax[0].plot(months, grouped['TEMPERATURA DO AR - BULBO SECO. HORARIA (C)'], label='Temperatura do Ar', color='tab:red', linewidth=2)
ax[0].set_title('Temperatura do Ar ao longo do tempo (média mensal)')
ax[0].set_ylabel('Temperatura (C)')
ax[0].legend()
ax[0].grid(True)

# Gráfico 2: Precipitação Total ao longo do tempo (média mensal)
ax[1].plot(months, grouped['PRECIPITAO TOTAL. HORARIO (mm)'], label='Precipitação Total', color='tab:blue', linewidth=2)
ax[1].set_title('Precipitação Total ao longo do tempo (média mensal)')
ax[1].set_xlabel('Data')
ax[1].set_ylabel('Precipitação (mm)')
ax[1].legend()
ax[1].grid(True)

# Ajustando apresentaçao
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
