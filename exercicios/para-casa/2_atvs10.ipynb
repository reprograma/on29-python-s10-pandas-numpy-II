```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data_sample_AM.csv', sep=';')
df
```

```python
df.describe()
```

```python
df[['PRECIPITAO TOTAL. HORARIO (mm)', 'TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)', 'TEMPERATURA MINIMA NA HORA ANT. (AUT) (C)']] = df[['PRECIPITAO TOTAL. HORARIO (mm)', 'TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)', 'TEMPERATURA MINIMA NA HORA ANT. (AUT) (C)']].fillna(df[['PRECIPITAO TOTAL. HORARIO (mm)', 'TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)', 'TEMPERATURA MINIMA NA HORA ANT. (AUT) (C)']].mean())
```

```python
df.isnull().sum() # verificando se eliminei os valores NaN
```

```python
df_colunas = df[['PRECIPITAO TOTAL. HORARIO (mm)', 'TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)', 'TEMPERATURA MINIMA NA HORA ANT. (AUT) (C)']]
df_colunas
```

```python
import matplotlib.pyplot as plt


df['Data'] = pd.to_datetime(df['Data'])
media_ano_temp= df[['TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)','TEMPERATURA MINIMA NA HORA ANT. (AUT) (C)']].groupby(df['Data'].dt.strftime('%m/%Y')).mean()

media_ano_temp.columns = [ 'Temperatura Máxima', 'Temperatura Mínima']

plt.title('Média Anual de Temperaturas\n(2020 - 2022)')
plt.xlabel('Ano')
plt.ylabel('Temperatura')

plt.grid()
plt.tick_params(top=False, bottom=False, left=False, right=False,
                labelleft=True, labelbottom=True)

#plt.plot(media_ano_temp, marker = 'o')

plt.plot(media_ano_temp['Temperatura Máxima'], marker='o', label='Temperatura Máxima')
plt.plot(media_ano_temp['Temperatura Mínima'], marker='o', label='Temperatura Mínima')

plt.xticks(rotation=90)
plt.legend()
```

```python
import matplotlib.pyplot as plt


df['Data'] = pd.to_datetime(df['Data'])
agrup_chuva = df[['PRECIPITAO TOTAL. HORARIO (mm)']].groupby(df['Data'].dt.strftime('%m/%Y')).agg(['mean', 'max', 'min'])

agrup_chuva.columns = ['Média de precipitação', 'Precipitação Máxima', 'Precipitação Mínima']

plt.title('Registro sobre precipitação em mm\n(2020 - 2022)')
plt.xlabel('Ano')
plt.ylabel('mm')

plt.grid()

plt.plot(agrup_chuva, marker = 'o')
plt.plot(agrup_chuva['Média de precipitação'], marker='o', label='Média de precipitação')
plt.plot(agrup_chuva['Precipitação Máxima'], marker='o', label='Precipitação Máxima')
plt.plot(agrup_chuva['Precipitação Mínima'], marker='o', label='Precipitação Mínima')

plt.xticks(rotation=90)
plt.legend()
```
