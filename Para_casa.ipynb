```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_sample_.csv', sep=';' )
```

```python
df
```

```python
df['REGIAO'] = df['REGIAO'].replace({'CO':'Centro-Oente','N':'Norte','NE':'Nordeste','S':'Sul','SE':'Suldeste'})
```

```python
df
```

```python
media_temperatura = df.groupby('REGIAO')['TEMPERATURA DO AR - BULBO SECO. HORARIA (C)'].mean()
```

```python
media_temperatura
```

```python

media_temperatura.plot(kind= 'bar')

plt.xlabel('Região')
plt.title('Análise da média da Temperatura por Região')
plt.ylabel('Temperatura (C)')

```

```python
df = df.dropna(subset = ['PRECIPITAO TOTAL. HORARIO (mm)','RADIACAO GLOBAL (Kj/m2)'])
```

```python
df
```

```python
normalizacao_radiacao = 'RADIACAO GLOBAL (Kj/m2)'
```

```python
normalizacao_temperatura = 'TEMPERATURA DO AR - BULBO SECO. HORARIA (C)'
```

```python
df[normalizacao_radiacao] = (df[normalizacao_radiacao] - df[normalizacao_radiacao].min())/(df[normalizacao_radiacao].max() - df[normalizacao_radiacao].min())
```

```python
df[normalizacao_temperatura] = (df[normalizacao_temperatura] - df[normalizacao_temperatura].min())/(df[normalizacao_temperatura].max() - df[normalizacao_temperatura].min())
```

```python
grafico = df.groupby('Hora UTC')[['RADIACAO GLOBAL (Kj/m2)','TEMPERATURA DO AR - BULBO SECO. HORARIA (C)']].mean().plot()
```
