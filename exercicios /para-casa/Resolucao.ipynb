```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime as dt
```

```python
df = pd.read_csv('data_sample_RS.csv', sep=';', encoding='latin-1', on_bad_lines='skip')
```

```python
df
```

```python
df.dropna()
```

# Gráfico 01
Criando gráfico para mostrar a média da temperatura por Hora.



```python
media_temperatura_hora = df[['Hora UTC', 'TEMPERATURA DO AR - BULBO SECO. HORARIA (C)']].groupby('Hora UTC').mean()
media_temperatura_hora.plot()
plt.xlabel('Horário')
plt.ylabel('Temperatura')
plt.title('Média de Temperatura por Hora')
```

# Gráfico 02
Velocidade Máxima de Vento por hora

```python
velocidade_max_vento = df[['Hora UTC', 'VENTO. RAJADA MAXIMA (m/s)']].groupby('Hora UTC').max()
velocidade_max_vento.plot()
plt.xlabel('Horário')
plt.ylabel('Velocidade')
plt.title('Velocidade máxima por Hora')
```
